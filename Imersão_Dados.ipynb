## 1.B - Separando os municípios por regiões com base nos dados do wikipedia

fonte: https://pt.wikipedia.org/wiki/Regi%C3%A3o_Metropolitana_de_S%C3%A3o_Paulo

fonte: https://pt.wikipedia.org/wiki/Litoral_de_S%C3%A3o_Paulo


```
municipios = dados['NO_MUNICIPIO_RESIDENCIA'].unique()
```

```
# separação por em quatro regiões para análise exploratória dos dados. 

sao_paulo_capital = []
sao_paulo_regiao_metropolitana = []
sao_paulo_litoral = []
sao_paulo_interior = []

capital = ['São Paulo']

metropolitana = ['Barueri' , 'Biritiba Mirim', 'Caieiras', 'Cajamar', 'Carapicuíba', 'Cotia', 
                 'Diadema', 'Embu das Artes', 'Embu-Guaçu', 'Ferraz de Vasconcelos', 'Francisco Morato',
                 'Franco da Rocha', 'Guararema', 'Guarulhos', 'Itapecerica da Serra', 'Itapevi', 'Itaquaquecetuba',
                 'Jandira', 'Juquitiba', 'Mairiporã', 'Mauá', 'Mogi das Cruzes', 'Osasco', 'Pirapora do Bom Jesus',
                 'Poá', 'Ribeirão Pires', 'Rio Grande da Serra', 'Salesópolis', 'Santa Isabel', 'Santana de Parnaíba',
                 'Santo André', 'São Bernardo do Campo', 'São Caetano do Sul', 'São Lourenço da Serra', 'Suzano',
                 'Taboão da Serra', 'Vargem Grande Paulista']

litoral = ['Santos', 'São Vicente', 'Praia Grande', 'Guarujá', 'Cubatão', 'Caraguatatuba', 'Itanhaém', 'Ubatuba',
           'São Sebastião', 'Peruíbe', 'Bertioga', 'Mongaguá', 'Ilhabela', 'Iguape', 'Cananéia', 'Ilha Comprida']

for municipio in municipios:

  if municipio in capital:
    sao_paulo_capital.append(municipio)

  elif municipio in metropolitana:
    sao_paulo_regiao_metropolitana.append(municipio)
  
  elif municipio in litoral:
    sao_paulo_litoral.append(municipio)

  else:
    sao_paulo_interior.append(municipio)  
```

# 1.C - Criando uma variável "REGIAO", inserindo os dados referenciais (Capital, Metropolitana, Interior e Litoral) e concatenado em um DataFrame único (Microdados_sp). 

```
capital = dados['NO_MUNICIPIO_RESIDENCIA'].isin(sao_paulo_capital)
dados_capital = dados[capital]
dados_capital['REGIAO'] = 'Capital'
```

```
interior = dados['NO_MUNICIPIO_RESIDENCIA'].isin(sao_paulo_interior)
dados_interior = dados[interior]
dados_interior['REGIAO'] = 'Interior'
```

```
metropolitana = dados['NO_MUNICIPIO_RESIDENCIA'].isin(sao_paulo_regiao_metropolitana)
dados_metropolitana = dados[metropolitana]
dados_metropolitana['REGIAO'] = 'Metropolitana'
```

```
litoral = dados['NO_MUNICIPIO_RESIDENCIA'].isin(sao_paulo_litoral)
dados_litoral = dados[litoral]
dados_litoral['REGIAO'] = 'Litoral'
```

```
microdados_sp = pd.concat([dados_capital,dados_interior,dados_litoral,dados_metropolitana], axis=0, ignore_index= True)
microdados_sp
```

```
microdados_sp['REGIAO'].value_counts()
```

## 5.A - Predição para Nota da prova de Linguagens e Códigos

-> Tentarei identificar o intervalo de confiança para a Nota da prova de Linguagens e Códigos.

```
provas_entrada = ["NU_NOTA_CH","NU_NOTA_MT", "NU_NOTA_CN","NU_NOTA_REDACAO"]
prova_saida = "NU_NOTA_LC"
dados_sem_notas_zero = dados_sem_notas_zero[provas].dropna()
notas_entrada = dados_sem_notas_zero[provas_entrada]
notas_saida = dados_sem_notas_zero[prova_saida]
```

```
x = notas_entrada
y = notas_saida 
```

```
from sklearn.model_selection import train_test_split

SEED = 231

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.20,
                                                        random_state=SEED)
```

```
from sklearn.svm import LinearSVR

modelo = LinearSVR(random_state = SEED)
modelo.fit(x_treino, y_treino)
```

```
predicoes_linguagem_codigo = modelo.predict(x_teste)
```

```
plt.figure(figsize=(8, 8))
sns.scatterplot(x=predicoes_linguagem_codigo, y=y_teste)
plt.xlim((-10, 850))
plt.ylim((-10, 850))
```

```
plt.figure(figsize=(8, 8))
sns.scatterplot(x=y_teste, y=x_teste.mean(axis=1))
plt.xlim((-10, 1000))
plt.ylim((-10, 1000))
```

```
resultados = pd.DataFrame()
resultados["Real"] = y_teste
resultados["Previsao"] = predicoes_linguagem_codigo
resultados["diferenca"] = resultados["Real"] - resultados["Previsao"]
resultados["quadrado_diferenca"] = (resultados["Real"] - resultados["Previsao"])**2
```

```
from sklearn.model_selection import cross_validate

modelo_arvore = DecisionTreeRegressor(max_depth=2)
resultados = cross_validate(modelo_arvore, x, y, cv= 10, scoring="neg_mean_squared_error")
media = (resultados["test_score"]*-1).mean()
desvio_padrao = (resultados["test_score"]*-1).std()
lim_inferior = media - (2*desvio_padrao)
lim_superior = media + (2*desvio_padrao)

print(f"Intervalo de confiança {lim_inferior} - {lim_superior}")
```
