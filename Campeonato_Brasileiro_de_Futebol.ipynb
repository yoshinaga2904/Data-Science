### Descrição:
<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>O Campeonato Brasileiro de Futebol, também conhecido como Campeonato Brasileiro, Brasileirão e Série A, é a liga brasileira de futebol profissional entre clubes do Brasil, sendo a principal competição futebolística no país.</p>

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>A fórmula de disputa do campeonato foi padronizada somente em 2003, quando foi adotado o sistema de pontos corridos com todas as equipes se enfrentando em turno e returno.</p>

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>O objetivo deste treinamento será estimar <b>Dados Estatísticos</b> utilizando técnicas de <b>Data Science</b> buscando projeções, curiosidades, estatísticas descritivas, estimações e meu desenvolvimento como cientísta de dados.</p>

<p style='font-size: 18px; line-height: 2; margin: 10px 50px; text-align: justify;'>Os dados (amostra) foram coletados da tabela de classificação final do campeonato brasileiro de futebol, dos anos de 2006 à 2019. O período levantado foi escolhido para manter um padrão pela quantidade de times participantes por campeonato, onde no ano de 2006 até 2019 foram disputados com 20 times em 38 rodadas.</p>

### Dados:
<ul style='font-size: 18px; line-height: 2; text-align: justify;'>
    <li><b>time</b> - clube de futebol</li>
    <li><b>ano</b> - Ano do campeonato disputado pelo time</li>
    <li><b>colocação</b> - Colocação final do time</li>
    <li><b>pontos</b> - Quantidade de pontos conquistados</li>
    <li><b>jogos</b> - Quantidade de jogos disputados</li>
    <li><b>vitorias</b> - Quantidade de vitorias conquistadas</li>
    <li><b>empates</b> - Quantidade de empates</li>
    <li><b>derrotas</b> - Quantidade de derrotas sofridas</li>
    <li><b>gols_pro</b> - Quantidade de gols marcados</li>
    <li><b>gols_contra</b> - Quantidade de gols sofridos</li>
    <li><b>saldo_gols</b> - Saldo de gols</li>
    <li><b>aproveitamento %</b> - Aproveitamento dos pontos em disputa</li>
</ul>

## Leitura dos dados

```
import pandas as pd
```

```
dados = pd.read_excel('brasileiro.xlsx')
dados.head()
```

```
dados['aproveitamento'] = ((dados['pontos'] / (dados['jogos']*3))*100).round(2)
dados.head()
```

```
santos = (dados['time'] == 'SANTOS')
quantidade = dados[santos].shape[0]
quantidade
```

```
classificacao = dados['time'] == 'SANTOS'
santos = dados[classificacao]
santos
```

```
santos.describe().round(2)
```

```
import numpy as np

frequencia = santos['colocacao'].value_counts()
percentual = santos['colocacao'].value_counts(normalize = True) * 100
distr_freq = pd.DataFrame({'Frequência': frequencia, 'Porcentagem (%)': percentual })
distr_freq
```

```
distr_freq.sort_index(ascending= True)
```

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
ax = sns.distplot(santos.colocacao)

ax.figure.set_size_inches(12, 6)
ax.set_title('Distribuição de Frequencias - Colocação', fontsize=18)
ax.set_xlabel('Quantidade', fontsize=14)
ax
```

```
ax = sns.boxplot(x= 'pontos', data= santos, orient= 'h')
ax.figure.set_size_inches(12, 4)
ax.set_title('Campeonato Brasileiro de 2006 a 2019', fontsize=18)
ax.set_xlabel('pontos', fontsize=14)
ax
```

```
frequencia = pd.crosstab(santos.colocacao, santos.pontos)

frequencia
```

```
tabela = dados['colocacao'] == 1
campeoes = dados[tabela]
campeoes
```

```
classif_santos = santos['aproveitamento'].max()
classif_santos
```

```
classif_santos = santos['aproveitamento'] == 64.91
melhor_santos = santos[classif_santos]
melhor_santos
```

```
projecao = pd.concat([campeoes, melhor_santos])
projecao.sort_values(by='aproveitamento', axis=0, ascending=True, inplace= True)
projecao.reset_index(inplace= True, drop= 'index')
projecao
```

```
fig = projecao['aproveitamento']
plt.figure(figsize=(15, 8))
fig.set_axis(labels=[projecao.time, projecao.ano], inplace= True)
fig.plot(kind='barh', title= 'Aproveitamento dos campeões brasileiros de 2006 à 2019 X melhor Santos FC 2019', fontsize= 14)
```
